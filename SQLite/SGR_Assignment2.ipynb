```python
filepath = '/content/drive/MyDrive/Data Management and Analytics/'

conn = sqlite3.connect(filepath+"IMDB_horror_movies.db")

# inspecting data
df = pd.read_sql_query('select * from IMDB_horror_movies', conn)
df
```

```python
def execSQL(conn,query):
  conn.execute(query)
  conn.commit()
```

1. Create a new table from the provided table and add a movie ID

```python
execSQL(conn,"drop table if exists movies")

execSQL(conn,"CREATE TABLE temp AS SELECT DISTINCT * FROM IMDB_horror_movies")
q='''
CREATE TABLE movies AS
SELECT ROWID as movie_id, *
  FROM temp
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists temp")

#checking new table
pd.read_sql_query("select * from movies",conn)
```

2. Move the genres to another table and add a genre ID

```python
execSQL(conn,"drop table if exists genres")

# creating table with unique genres (one genre per row)
df = pd.read_sql_query("SELECT genres FROM movies",conn)
genres_temp = df.assign(genres=df.genres.str.split('|')).explode('genres')
genres_temp.drop_duplicates().to_sql('genres_temp', conn)

q = '''
CREATE TABLE genres AS
  SELECT ROWID as genre_id, genres
  FROM genres_temp
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists genres_temp")

# dropping genres from movies
execSQL(conn,'''ALTER TABLE movies DROP COLUMN genres''')

# checking new genres table
pd.read_sql_query("select * from genres", conn)
```

3. Create a table which contains the relationships between the movies and the genres
(using IDs only). A given movie can have more than one genre.

```python
execSQL(conn,"drop table if exists movies_genres")

# creating table with movie_id and genres linked together
genres_temp['movie_id'] = genres_temp.index + 1
genres_temp.to_sql('movie_id_genres', conn)

# joining tables to get movie_id and genre_id
q = '''
CREATE TABLE movies_genres AS
SELECT a.movie_id, b.genre_id
FROM movie_id_genres as a LEFT JOIN genres as b ON a.genres=b.genres
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists movie_id_genres")

# checking new table (making it type int since None converted it to float)
pd.read_sql_query("select * from movies_genres", conn).astype("Int64")
```

4. Move the release countries to another table, add a country ID, and replace the release_country in the table of movies with the country ID as a foreign key. 

```python
execSQL(conn,"drop table if exists countries")

execSQL(conn,"CREATE TABLE temp AS SELECT DISTINCT release_country FROM movies")
q='''
CREATE TABLE countries AS
SELECT ROWID as country_id, release_country
  FROM temp
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists temp")

#checking new table
pd.read_sql_query("select * from countries",conn)
```

```python
# replacing release_country with country_id in movies as foreign key
execSQL(conn,"drop table if exists movies_new")

q='''
CREATE TABLE movies_new AS
  SELECT movies.movie_id, movies.title, movies.release_date, countries.country_id,
        movies.movie_rating, movies.review_rating, movies.movie_runtime, movies.plot, movies.movie_cast,
        movies.movie_language, movies.filming_locations, movies.budget
  FROM movies JOIN countries USING(release_country)
'''
execSQL(conn, q)

# checking updated movies table 
pd.read_sql_query("select * from movies_new",conn)
```

```python
# replacing movies table with updated version

execSQL(conn,"drop table if exists movies")
execSQL(conn,'''ALTER TABLE movies_new RENAME TO movies''')
```

6. Create a table of movie locations with location ID and add the location ID as a foreign key to the movie table.

```python
execSQL(conn,"drop table if exists locations")

# handling missing values
execSQL(conn,"CREATE TABLE temp AS SELECT DISTINCT ifnull(filming_locations, 'None') AS filming_locations FROM movies")

q='''
CREATE TABLE locations AS
SELECT ROWID as location_id, filming_locations
  FROM temp
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists temp")

#checking new table
pd.read_sql_query("select * from locations",conn)
```

```python
# replacing filming_locations with location_id in movies as foreign key
execSQL(conn,"drop table if exists movies_new")

q='''
CREATE TABLE movies_new AS
  SELECT movies.movie_id, movies.title, movies.release_date, movies.country_id,
        movies.movie_rating, movies.review_rating, movies.movie_runtime, movies.movie_cast,
        movies.movie_language, locations.location_id, movies.budget
  FROM movies LEFT JOIN locations ON ifnull(movies.filming_locations, 'None') == locations.filming_locations
'''
execSQL(conn, q)

# checking updated movies table 
pd.read_sql_query("select * from movies_new",conn)
```

```python
# replacing movies table with updated version

execSQL(conn,"drop table if exists movies")
execSQL(conn,'''ALTER TABLE movies_new RENAME TO movies''')
```

7. Move the languages to another table and add a language ID. 

```python
execSQL(conn,"drop table if exists languages")

# creating table with unique languages (one language per row)
df = pd.read_sql_query("SELECT movie_language FROM movies",conn)
lang_temp = df.assign(movie_language=df.movie_language.str.split('|')).explode('movie_language')
lang_temp.drop_duplicates().to_sql('languages_temp', conn)

q = '''
CREATE TABLE languages AS
  SELECT ROWID as language_id, movie_language
  FROM languages_temp
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists languages_temp")

# dropping movie_language from movies
execSQL(conn,'''ALTER TABLE movies DROP COLUMN movie_language''')

# checking new languages table
pd.read_sql_query("select * from languages", conn)
```

8. Create a table which contains the relationships between the movies and the languages (using IDs only).

```python
execSQL(conn,"drop table if exists movies_languages")

# creating table with movie_id and language linked together
lang_temp['movie_id'] = lang_temp.index + 1
lang_temp.to_sql('movie_id_lang', conn)

# joining tables to get movie_id and language_id
q = '''
CREATE TABLE movies_languages AS
SELECT a.movie_id, b.language_id
FROM movie_id_lang as a LEFT JOIN languages as b ON a.movie_language=b.movie_language
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists movie_id_lang")

# checking new table
pd.read_sql_query("select * from movies_languages", conn).astype("Int64")
```

9. Remove all unique actors that are in the movie cast to their own table which has individual actor names and an actor ID. Delete the movie cast column.

```python
execSQL(conn,"drop table if exists actors")

# creating table with unique actors (one actor per row)
df = pd.read_sql_query("SELECT movie_cast FROM movies",conn)
act_temp = df.assign(movie_cast=df.movie_cast.str.split('|')).explode('movie_cast')
act_temp.drop_duplicates().to_sql('actors_temp', conn)

q = '''
CREATE TABLE actors AS
  SELECT ROWID as actor_id, movie_cast
  FROM actors_temp
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists actors_temp")

# dropping movie_cast from movies
execSQL(conn,'''ALTER TABLE movies DROP COLUMN movie_cast''')

# checking new actors table
pd.read_sql_query("select * from actors", conn)
```

10. Represent the movie-cast relationship as a table that consists of the movie id and actor ID.

```python
execSQL(conn,"drop table if exists movies_actors")

# creating table with movie_id and actors linked together
act_temp['movie_id'] = act_temp.index + 1
act_temp.to_sql('movie_id_act', conn)

# joining tables to get movie_id and act_id
q = '''
CREATE TABLE movies_actors AS
SELECT a.movie_id, b.actor_id
FROM movie_id_act as a LEFT JOIN actors as b ON a.movie_cast=b.movie_cast
'''
execSQL(conn,q)
execSQL(conn,"drop table if exists movie_id_act")

# checking new table
pd.read_sql_query("select * from movies_actors", conn).astype("Int64")
```
