```python
import twint
import pandas as pd
import nest_asyncio
from datetime import date, timedelta
nest_asyncio.apply()
```

```python
# testing twint

c = twint.Config()

c.Store_object = True
c.Pandas = True
c.Search='loneliness COVID-19'
c.Since='2020-05-01'
c.Until= '2020-05-02'
c.Lang = 'en'
twint.run.Search(c)

twint.storage.panda.Tweets_df
```

```python
Corpus = pd.DataFrame()

start_date = date(2020, 5, 1) 
end_date = date(2020, 6, 30)
delta = end_date - start_date

c = twint.Config()
c.Store_object = True
c.Pandas = True
c.Search='loneliness COVID-19'
c.Lang = 'en'

for i in range(delta.days + 1):
  since = str(start_date + timedelta(days=i))
  until = str(start_date + timedelta(days=i+1))

  c.Since = since
  c.Until = until
  twint.run.Search(c)

  Corpus = pd.concat([Corpus, twint.storage.panda.Tweets_df.drop_duplicates(subset=['id'])])
```

##**Exporting the Data**

```python
Corpus.index = range(len(Corpus))

Corpus.to_csv('covid_loneliness_tweets.csv', index = False)
```
